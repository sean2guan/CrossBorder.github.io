```python
import pandas as pd
df_url = pd.read_csv('url.csv')
df_parent_categories = pd.read_csv('parent_categories.csv')
df_child_categories = pd.read_csv('child_categories.csv')
```

```python
main_menu_whole = ''
for index, row in df_parent_categories.iterrows():
    main_menu = '''
                    <li>
                        <a>
                            <i class="linecons-lightbulb"></i>
                            <span class="title">%s</span>
                        </a>
                        <ul>'''%(row['parent_categories'])
    for index, row in df_child_categories[(df_child_categories['parent_categories']==row['parent_categories'])].iterrows():
        content_child_categories = '''
                            <li>
                                <a href="#%s" class="smooth">
                                    <span class="title">%s</span>
                                </a>
                            </li>'''%(row['child_categories'],row['child_categories'])
        main_menu = main_menu + content_child_categories


    main_menu = main_menu + '''
                        </ul>
                    </li>'''  
    main_menu_whole =  main_menu_whole + main_menu
# print(main_menu_whole)
```

```python
main_content_whole = ''
for index, row in df_child_categories.iterrows():
    main_content = '''
            <h4 class="text-gray"><i class="linecons-tag" style="margin-right: 7px;" id="%s"></i>%s</h4>'''%(row['child_categories'],row['child_categories'])
    
    df = df_url[(df_url['categories']==row['child_categories'])]
    for index, row in df.reset_index(drop=True).iterrows():
        if index%4==0 :
            main_content = main_content + '''
            <div class="row">'''        
        content_url = '''
                <div class="col-sm-3">
                    <div class="xe-widget xe-conversations box2 label-info" onclick="window.open('%s', '_blank')" data-toggle="tooltip" data-placement="bottom" title="" data-original-title="%s">
                        <div class="xe-comment-entry">
                            <a class="xe-user-img">
                                <img data-src="../assets/images/logos/%s" class="lozad img-circle" width="40">
                            </a>
                            <div class="xe-comment">
                                <a href="#" class="xe-user-name overflowClip_1">
                                    <strong>%s</strong>
                                </a>
                                <p class="overflowClip_2">%s</p>
                            </div>
                        </div>
                    </div>
                </div>'''%(row['url'],row['url'],row['icon'],row['name'],row['description'])
        main_content = main_content + content_url
        if index%4==3 or index==len(df)-1:
            main_content = main_content + '''
            </div>'''
#         print(len(df))
#         print(row)
    main_content = main_content + '''
            <br />'''   
    main_content_whole = main_content_whole + main_content
# print(main_content_whole)
```
